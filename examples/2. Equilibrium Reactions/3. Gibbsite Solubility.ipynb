# GIBBSITE SOLUBILITY AS A FUNCTION OF PH
http://hydrochemistry.eu/exmpls/gibbsite.html

```python
%pylab inline
import phreeqpython
pp = phreeqpython.PhreeqPython('phreeqc.dat')
```

```python
x = []
y = []

for ph in np.linspace(3,12, 17):
    sol = pp.add_solution({'pH': ph, 'Al': '1e3 Gibbsite'})
    x.append(ph)
    y.append(sol.total_element('Al', units='mol'))
```

```python
plt.figure(figsize=[10,5])
plt.plot(x, y, 'rs-')
plt.title('Gibbsite Equilibrium')
plt.xlim(0,14)
plt.yscale('log')
plt.grid()
```
